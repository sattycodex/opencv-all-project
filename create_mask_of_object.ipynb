```python
import cv2
import numpy as np
```

```python
def nothing(x):
    pass


cv2.namedWindow('maskable')
switch='0:OFF|1:ON'
cv2.createTrackbar(switch,'maskable',0,1,nothing)
cv2.createTrackbar('lh','maskable',0,255,nothing)
cv2.createTrackbar('ls','maskable',0,255,nothing)
cv2.createTrackbar('ll','maskable',0,255,nothing)
cv2.createTrackbar('hh','maskable',255,255,nothing)
cv2.createTrackbar('hs','maskable',255,255,nothing)
cv2.createTrackbar('hl','maskable',255,255,nothing)


cap=cv2.VideoCapture(0)
while (True):
    ret,frame=cap.read()
    if ret==False:
        break
    else:
        hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
        s=cv2.getTrackbarPos(switch,'maskable')
        lh=cv2.getTrackbarPos('lh','maskable')
        ls=cv2.getTrackbarPos('ls','maskable')
        ll=cv2.getTrackbarPos('ll','maskable')
        hh=cv2.getTrackbarPos('hh','maskable')
        hs=cv2.getTrackbarPos('hs','maskable')
        hl=cv2.getTrackbarPos('hl','maskable')
        
        if s==0:
            lower=np.array([0,0,0])
            upper=np.array([0,0,0])
            
        else:
            lower=np.array([lh,ls,ll])
            upper=np.array([hh,hs,hl])
            
       
        mask = cv2.inRange(hsv, lower, upper)
        useful=cv2.bitwise_and(frame,frame,mask=mask)
        cv2.imshow('video',frame)
        cv2.imshow('video1',mask)
        cv2.imshow('video3',useful)   
        
        
        
        if cv2.waitKey(1) & 0xFF==ord('q'):
            break
            
cap.release()
cv2.destroyAllWindows()
```

```python
frame=cv2.imread('image.jpg')

def nothing(x):
    pass


cv2.namedWindow('maskable')
cv2.resizeWindow('maskable',(400,500))
switch='0:OFF|1:ON'
cv2.createTrackbar(switch,'maskable',0,1,nothing)
cv2.createTrackbar('lh','maskable',0,255,nothing)
cv2.createTrackbar('ls','maskable',0,255,nothing)
cv2.createTrackbar('ll','maskable',0,255,nothing)
cv2.createTrackbar('hh','maskable',255,255,nothing)
cv2.createTrackbar('hs','maskable',255,255,nothing)
cv2.createTrackbar('hl','maskable',255,255,nothing)

while (True):
    
        hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
        s=cv2.getTrackbarPos(switch,'maskable')
        lh=cv2.getTrackbarPos('lh','maskable')
        ls=cv2.getTrackbarPos('ls','maskable')
        ll=cv2.getTrackbarPos('ll','maskable')
        hh=cv2.getTrackbarPos('hh','maskable')
        hs=cv2.getTrackbarPos('hs','maskable')
        hl=cv2.getTrackbarPos('hl','maskable')
        
        if s==0:
            lower=np.array([0,0,0])
            upper=np.array([0,0,0])
            
        else:
            lower=np.array([lh,ls,ll])
            upper=np.array([hh,hs,hl])
            
       
        mask = cv2.inRange(hsv, lower, upper)
        useful=cv2.bitwise_and(frame,frame,mask=mask)
        frame=cv2.resize(frame,(400,500))
        mask=cv2.resize(mask,(400,500))
        useful=cv2.resize(useful,(400,500))
        cv2.imshow('video',frame)
        cv2.imshow('video1',mask)
        cv2.imshow('video3',useful)   
        
        
        
        if cv2.waitKey(1) & 0xFF==ord('q'):
            break
            
cap.release()
cv2.destroyAllWindows()
```
